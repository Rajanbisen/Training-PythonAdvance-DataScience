```python
fig, ax = plt.subplots()

ax.plot(a, a+1, color="blue", alpha=0.5)
ax.plot(a, a+2, color="#8B008B")        
ax.plot(a, a+3, color="#FF8C00")        
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(a, a+1, color="red", linewidth=0.25)
ax.plot(a, a+2, color="red", linewidth=0.50)
ax.plot(a, a+3, color="red", linewidth=1.00)
ax.plot(a, a+4, color="red", linewidth=2.00)

# possible linestype options ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(a, a+5, color="green", lw=3, linestyle='-')
ax.plot(a, a+6, color="green", lw=3, ls='-.')
ax.plot(a, a+7, color="green", lw=3, ls=':')

# custom dash
line, = ax.plot(a, a+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(a, a+ 9, color="blue", lw=3, ls='-', marker='+')
ax.plot(a, a+10, color="blue", lw=3, ls='--', marker='o')
ax.plot(a, a+11, color="blue", lw=3, ls='-', marker='s')
ax.plot(a, a+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
ax.plot(a, a+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(a, a+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(a, a+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(a, a+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].plot(a, a**2, a, a**3)
axes[0].set_title("default axes ranges")

axes[1].plot(a, a**2, a, a**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(a, a**2, a, a**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```
